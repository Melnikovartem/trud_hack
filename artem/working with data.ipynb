```python
#готовим фичи
data = prof.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
dd = [0 for i in range(len(set(data["ТЕКУЩАЯ ЗАНЯТОСТЬ"])))]
for i in data["ТЕКУЩАЯ ЗАНЯТОСТЬ"]:
    dd[i] += 1
max(dd)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     data.drop(columns = "ТЕКУЩАЯ ЗАНЯТОСТЬ"), data["ТЕКУЩАЯ ЗАНЯТОСТЬ"], test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier


model = KNeighborsClassifier(n_neighbors = 5, metric='minkowski', p=2)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
len(y_pred) == len(y_test)
```
