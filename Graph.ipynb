```python
import ParsFromVK
import pymorphy2
import string

def check_Gramme(word, morth=pymorphy2.MorphAnalyzer()): # –ì—Ä–∞–º–º–µ–º–∞ —Å–ª–æ–≤–∞
    
    return morth.parse(word)[0].tag.POS


def get_normal_form(word, morth=pymorphy2.MorphAnalyzer()): # –ù–æ—Ä–º–∞–ª—å–Ω–∞—è —Ñ–æ—Ä–º–∞ —Å–ª–æ–≤–∞
    
    return morth.parse(word)[0].normal_form


def delete_punctuation(word): # –£–¥–∞–ª–µ–Ω–∏–µ –≤—Å–µ—Ö –∑–Ω–∞–∫–æ–≤ –ø—É–Ω–∫—Ç—É–∞—Ü–∏–∏
    
    tt = str.maketrans(dict.fromkeys(string.punctuation))
    
    return word.translate(tt)


forbidden_grammes = {'NPRO', 'PRCL', 'INTJ', 'CONJ', 'PREP', 'PRED', 'PNCT'}  # –ó–∞–ø—Ä–µ—â–µ–Ω–Ω—ã–µ –≥—Ä–∞–º–º–µ–º—ã

print(get_normal_form('–ø–∞—Ä–Ω–∏!'), check_Gramme('–ø–∞—Ä–Ω–∏'))

number_vertex = {} # –ù–æ–º–µ—Ä –≤–µ—Ä—à–∏–Ω—ã –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é
vertex_edges = {} # –°–æ—Å–µ–¥–∏ –≤–µ—Ä—à–∏–Ω—ã

num = 1

# –ê–≥—Ä–æ–Ω–æ–º—ã

name_1 = '–ê–≥—Ä–æ–Ω–æ–º'
number_vertex[name_1] = num
num += 1
cnt = {} # —á–∏—Å–ª–æ —Ä–∞–∑ –∫–æ—Ç–æ—Ä–æ–µ —Å–ª–æ–≤–æ –≤—Å—Ç—Ä–µ—Ç–∏–ª–æ—Å—å
file = open('–ê–≥—Ä–æ–Ω–æ–º—ã.txt', 'r')
for line in file:
    cur_id = line
    main_information, b, groups = ParsFromVK.pars_from_vk(cur_id)
    for group in groups:
        words = group.split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                else:
                    cnt[word] += 1
    for key in main_information:
        words = str(main_information[key]).split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                else:
                    cnt[word] += 1
temp = []
for key in cnt:
    temp.append((key, cnt[key]))
print(temp)
```
