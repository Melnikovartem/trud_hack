```python
import ParsFromVK
import pymorphy2
import string

def check_Gramme(word, morth=pymorphy2.MorphAnalyzer()): # Граммема слова
    
    return morth.parse(word)[0].tag.POS


def get_normal_form(word, morth=pymorphy2.MorphAnalyzer()): # Нормальная форма слова
    
    return morth.parse(word)[0].normal_form


def delete_punctuation(word): # Удаление всех знаков пунктуации
    
    tt = str.maketrans(dict.fromkeys(string.punctuation))
    
    return word.translate(tt)


forbidden_grammes = {'NPRO', 'PRCL', 'INTJ', 'CONJ', 'PREP', 'PRED', 'PNCT'}  # Запрещенные граммемы
forbidden_words = {'title', 'True', 'False', 'id', 'false', 'true'}

print(get_normal_form('18121990'), check_Gramme('18121990'))

number_vertex = {} # Номер вершины по названию
vertex_edges = {} # Соседи вершины

num = 1

# Агрономы

name_1 = 'Агроном'
number_vertex[name_1] = num
vertex_edges[name_1] = []
num += 1
cnt = {} # число раз которое слово встретилось
file = open('Агрономы.txt', 'r')
tempcnt = 0
for line in file:
    tempcnt += 1
    if tempcnt >= 20:
        break
    cur_id = line
    if ParsFromVK.is_close(cur_id) == True:
        continue
    main_information, groups = ParsFromVK.pars_from_vk(cur_id)
    for group in groups:
        words = group.split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                else:
                    cnt[word] += 1
    for key in main_information:
        words = str(main_information[key]).split()
        for word in words:
            word = delete_punctuation(word)
            if check_Gramme(word) not in forbidden_grammes and check_Gramme(word) != None:
                word = get_normal_form(word)
                if word not in cnt:
                    cnt[word] = 1
                else:
                    cnt[word] += 1
temp = []
for key in cnt:
    if len(key) != 0 and key not in forbidden_words:
        temp.append((key, cnt[key]))
temp.sort(key=lambda point: (-point[1]))
for i in range(0, 50):
    curName = temp[i][0]
    curCnt = temp[i][1]
    print(curName, curCnt)
    if curName not in number_vertex:
        number_vertex[curName] = num
        num += 1
        vertex_edges[curName] = []
    vertex_edges[curName].append((name_1, curCnt))
    vertex_edges[name_1].append((curName, curCnt))
```
